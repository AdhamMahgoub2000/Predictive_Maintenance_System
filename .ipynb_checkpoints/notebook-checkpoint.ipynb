# Deep Learning–Based Predictive Maintenance System

## Load Dataset & Prepare

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler

def load_data(train_path):
    column_names = [
    "engine_id", "cycle", 
    "op_set_1", "op_set_2", "op_set_3"
    ] + [f"sensor_{i}" for i in range(1, 22)]  # 21 sensors in FD001

    train_df = pd.read_csv(
        train_path, 
        sep=r'\s+',  # handles multiple spaces
        header=None,
        names=column_names
        )
    train_df = train_df.drop(columns=train_df.columns[26:])  # remove empty extra columns
    return train_df
```

```python
train_path = "data/train_FD001.txt"
train_df   = load_data(train_path)
train_df.describe()
```

```python
def add_labels_train(train_df):
    train_df['RUL'] = train_df.groupby('engine_id')['cycle'].transform("max") - train_df['cycle']
    return train_df
```

```python
train_df = add_labels_train(train_df)
```

```python
test_path = "data/test_FD001.txt"
test_df = load_data(test_path)
test_df.describe()
```

```python
rul_path = "data/RUL_FD001.txt"
def load_rul_test(rul_path):
    rul_df = pd.read_csv(rul_path, header=None, names=["RUL"])
    rul_df['engine_id'] = test_df['engine_id'].unique()
    rul_df = rul_df[['engine_id', 'RUL']]
    return rul_df
```

```python
rul_path = "data/RUL_FD001.txt"
rul_df = load_rul_test(rul_path)
rul_df
```

### Creating Sliding Window

```python
import numpy as np

def create_sliding_windows(
    df,
    feature_cols,
    label_col,
    window_size,
    stride=1
    ):
    X, y = [], []

    for engine_id, engine_df in df.groupby('engine_id'):
        engine_df = engine_df.sort_values('cycle')

        features = engine_df[feature_cols].values
        labels = engine_df[label_col].values

        for i in range(0, len(engine_df) - window_size + 1, stride):
            X.append(features[i:i + window_size])
            y.append(labels[i + window_size - 1])  # label at last timestep

    return np.array(X), np.array(y)

def create_test_windows(
    df,
    feature_cols,
    window_size
):
    X = []
    engine_ids = []

    for engine_id, engine_df in df.groupby('engine_id'):
        engine_df = engine_df.sort_values('cycle')

        if len(engine_df) < window_size:
            continue  # or pad if you want

        window = engine_df[feature_cols].values[-window_size:]
        X.append(window)
        engine_ids.append(engine_id)

    return np.array(X), engine_ids
```

```python
feature_cols = [
    c for c in train_df.columns
    if c not in ['engine_id', 'cycle', 'label', 'RUL']
]
WINDOW_SIZE = 50

X_train, y_train = create_sliding_windows(
    train_df,
    feature_cols=feature_cols,
    label_col='label',
    window_size=WINDOW_SIZE
)
print(X_train.shape)
```

    create window for the last cycle in each engine in the test set

```python
X_test, test_engine_ids = create_test_windows(
    test_df,
    feature_cols=feature_cols,
    window_size=WINDOW_SIZE
)
rul_df
```

```python
def create_y_test(X_test, test_engine_ids, rul_df, label_col='label'):
  
    y_test = []

    for engine_id in test_engine_ids:
        engine_row = rul_df[rul_df['engine_id'] == engine_id]
        if len(engine_row) == 0:
            raise ValueError(f"Engine ID {engine_id} not found in rul_df")
        
        last_label = engine_row[label_col].values[-1]  # last row for this engine
        y_test.append(last_label)

    return np.array(y_test)
```

```python
y_test = create_y_test(X_test, test_engine_ids, rul_df, label_col='label')
print(X_train.shape, len(y_train))
print(X_test.shape, len(y_test))
```

## DATA LOADER

```python
import torch
from torch.utils.data import Dataset, DataLoader
import torch.nn as nn
import torch.optim as optim

class SequenceDataset(Dataset):
    def __init__(self, X, y):
        # Convert data to torch tensors
        self.X = torch.tensor(X, dtype=torch.float32)  # shape: (num_samples, seq_len, features)
        self.y = torch.tensor(y, dtype=torch.long)     # shape: (num_samples,)
    
    def __len__(self):
        return len(self.X)
    
    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]

# Example usage
train_dataset = SequenceDataset(X_train, y_train)
test_dataset = SequenceDataset(X_test, y_test)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)
```

## Develop the Model

```python
class PredictiveMaintenance(nn.Module):
    def __init__(self, input_size, cnn_channels, lstm_hidden_size, lstm_layers, num_classes, kernel_size=3):
        super(PredictiveMaintenance, self).__init__()
        
        # 1D CNN layer
        self.cnn = nn.Sequential(
            nn.Conv1d(in_channels=input_size, out_channels=cnn_channels, kernel_size=kernel_size, padding=kernel_size//2),
            nn.ReLU(),
            nn.Conv1d(in_channels=cnn_channels, out_channels=cnn_channels, kernel_size=kernel_size, padding=kernel_size//2),
            nn.ReLU()
        )
        
        # LSTM layer
        self.lstm = nn.LSTM(
            input_size=cnn_channels, 
            hidden_size=lstm_hidden_size, 
            num_layers=lstm_layers, 
            batch_first=True
        )
        
        # Fully connected layer
        self.fc = nn.Linear(lstm_hidden_size, num_classes)
    
    def forward(self, x):
        # x: (batch_size, seq_len, features)
        x = x.permute(0, 2, 1)  # -> (batch_size, features, seq_len) for Conv1d
        x = self.cnn(x)
        x = x.permute(0, 2, 1)  # -> (batch_size, seq_len, cnn_channels) for LSTM
        lstm_out, _ = self.lstm(x)  # lstm_out: (batch_size, seq_len, hidden_size)
        out = self.fc(lstm_out[:, -1, :])  # take last time step
        return out

# Example usage
input_size = X_train.shape[2]  # number of features
cnn_channels = 64
lstm_hidden_size = 128
lstm_layers = 1
num_classes = 3  

model = PredictiveMaintenance(input_size, cnn_channels, lstm_hidden_size, lstm_layers, num_classes)
```

```python
# Define the loss function
criterion = nn.CrossEntropyLoss()

# Define the optimizer
optimizer = optim.Adam(model.parameters(), lr=0.001)  
```

## Train The Model

```python
if torch.backends.mps.is_available() and torch.backends.mps.is_built():
    device = torch.device("mps")
else:
    device = torch.device("cpu")  # fallback to CPU

print(f"Using device: {device}")

# Move model to device
model = model.to(device)

# Training parameters
num_epochs = 20

for epoch in range(num_epochs):
    model.train()  # training mode
    running_loss = 0.0
    correct = 0
    total = 0
    
    for X_batch, y_batch in train_loader:
        X_batch = X_batch.to(device)
        y_batch = y_batch.to(device)
        
        optimizer.zero_grad()
        outputs = model(X_batch)
        loss = criterion(outputs, y_batch)
        loss.backward()
        optimizer.step()
        
        running_loss += loss.item() * X_batch.size(0)
        _, predicted = torch.max(outputs, 1)
        total += y_batch.size(0)
        correct += (predicted == y_batch).sum().item()
    
    train_loss = running_loss / total
    train_acc = correct / total
    
    # Evaluation
    model.eval()
    test_loss = 0.0
    correct = 0
    total = 0
    with torch.no_grad():
        for X_batch, y_batch in test_loader:
            X_batch = X_batch.to(device)
            y_batch = y_batch.to(device)
            
            outputs = model(X_batch)
            loss = criterion(outputs, y_batch)
            
            test_loss += loss.item() * X_batch.size(0)
            _, predicted = torch.max(outputs, 1)
            total += y_batch.size(0)
            correct += (predicted == y_batch).sum().item()
    
    test_loss /= total
    test_acc = correct / total
    
    print(f"Epoch [{epoch+1}/{num_epochs}] "
          f"Train Loss: {train_loss:.4f}, Train Acc: {train_acc:.4f} | "
          f"Test Loss: {test_loss:.4f}, Test Acc: {test_acc:.4f}")
```

## Evaluating Metrcis


```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

def evaluate_model_metrics(model, data_loader, criterion, device):
    """
    Evaluates a PyTorch model on a dataset across multiple metrics.

    Args:
        model: trained PyTorch model
        data_loader: DataLoader for dataset
        criterion: loss function (e.g., nn.CrossEntropyLoss)
        device: torch.device ('mps', 'cuda', 'cpu')

    Returns:
        metrics: dictionary with loss, accuracy, precision, recall, f1, confusion matrix
        all_preds: numpy array of predictions
        all_labels: numpy array of true labels
    """
    model.eval()
    test_loss = 0.0
    all_preds = []
    all_labels = []

    with torch.no_grad():
        for X_batch, y_batch in data_loader:
            X_batch = X_batch.to(device)
            y_batch = y_batch.to(device)
            
            outputs = model(X_batch)
            loss = criterion(outputs, y_batch)
            
            test_loss += loss.item() * X_batch.size(0)
            _, predicted = torch.max(outputs, 1)
            
            all_preds.extend(predicted.cpu().numpy())
            all_labels.extend(y_batch.cpu().numpy())
    
    all_preds = np.array(all_preds)
    all_labels = np.array(all_labels)
    
    test_loss /= len(all_labels)
    acc = accuracy_score(all_labels, all_preds)
    precision = precision_score(all_labels, all_preds, average='weighted', zero_division=0)
    recall = recall_score(all_labels, all_preds, average='weighted', zero_division=0)
    f1 = f1_score(all_labels, all_preds, average='weighted', zero_division=0)
    cm = confusion_matrix(all_labels, all_preds)
    
    metrics = {
        "loss": test_loss,
        "accuracy": acc,
        "precision": precision,
        "recall": recall,
        "f1_score": f1,
        "confusion_matrix": cm
    }
    
    print(f"Loss: {test_loss:.4f}, Accuracy: {acc:.4f}")
    print(f"Precision: {precision:.4f}, Recall: {recall:.4f}, F1-score: {f1:.4f}")
    print("Confusion Matrix:")
    print(cm)
    print("\nClassification Report:")
    print(classification_report(all_labels, all_preds, digits=4, zero_division=0))
    
    return metrics, all_preds, all_labels
_,_,_,= evaluate_model_metrics(model, test_loader, criterion, device)
```
